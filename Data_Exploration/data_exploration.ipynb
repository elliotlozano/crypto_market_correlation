```python
#import libraries and dependencies
import pandas as pd
from pathlib import Path
%matplotlib inline
```

```python
# import SPY daily returns
spy_daily_returns = pd.read_csv(Path('../Resources/spy_daily_returns.csv'))

# Create column headers and assign to DataFrame
columns = ['Date', "SPY Daily Return"]
spy_daily_returns.columns = columns

# Set Date as index
spy_daily_returns = spy_daily_returns.set_index('Date')

# Confirm header
spy_daily_returns.head()
```

```python
# plot SPY daily returns
spy_daily_returns.plot(
    figsize = (15,10),
    title = 'SPY - Daily Returns'
)
```

```python
# Create new DataFrame of cummulative returns
spy_cummulative_returns = (1 + spy_daily_returns).cumprod()

# Verify function worked by reviewing last few lines of DataFrame
spy_cummulative_returns.tail()
```

```python
# plot SPY cummulative returns
spy_cummulative_returns.plot(
    figsize = (15,10),
    title = 'SPY - Cummulative Returns'
)
```
